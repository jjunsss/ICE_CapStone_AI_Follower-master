```python
import socket
from jetbot import Robot
import time

robot = Robot()
```

```python
ADDRESS = "192.168.0.16"
PORT = 6215
```

```python
sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
sock.bind((ADDRESS, PORT))
LEFT_SPEED = 0.4
RIGHT_SPEED = 0.4

while True:
    data, addr = sock.recvfrom(1024)
    if data.decode() is not None:
        splitData = data.decode().split('|')
        if len(splitData) >= 3:
            if splitData[0] == 'left':
                print(f"Direction: {splitData[0]} speed: {splitData[1]}")
                robot.set_motors(LEFT_SPEED * float(splitData[1]), RIGHT_SPEED * float(splitData[2]))

            elif splitData[0] == 'right':
                print(f"Direction: {splitData[0]} speed: {splitData[1]}")
                robot.set_motors(LEFT_SPEED * float(splitData[1]), RIGHT_SPEED * float(splitData[2]))

            elif splitData[0] == 'front':
                print(f"Direction: {splitData[0]} speed: {splitData[1]}")
                robot.set_motors(LEFT_SPEED * float(splitData[1]), RIGHT_SPEED * float(splitData[2]))
            else:
                print("stop")
                robot.stop()

sock.close()\
```
